```python
#loading the hand written data sets
import pandas as pd
from sklearn.datasets import load_digits
```

```python
# Loading the handwritten digits dataset
digits = load_digits()
# Creating a Pandas DataFrame
df = pd.DataFrame(data=digits.data)
```

```python
# Performing EDA on the data. Show the shape, data types, first five lines.
# Ensure there are no missing values.
# 2.1 shape
df.shape
```

```python
# Data types
df.info()
```

```python
# first five lines.
df.head(5)
```

```python
# Showing all columns. Are the names useful? What would useful column names be in this case?
# Display all columns in the DataFrame
print(df.columns.values)
```

```python
#ensure no null data.
df.isnull().sum()
```

```python
# perform standardization by importing necessary libraries
# Importing and apply the ScandardScaler. Save the scaled datafrmae to "df_standardized"
from sklearn.preprocessing import StandardScaler
# Creating x & y Values
X = digits.data
Y = digits.target
# Creating an instance of the StandardScaler
scaler = StandardScaler()
# Fitting the scaler to our data
X_standardized = scaler.fit_transform(X)
df_standardized = pd.DataFrame(X_standardized)
```

```python
# Display (part of) the original and standardized dataframes.
# Printing First 3 rows of the original DataFrame
print("Original DataFrame:")
print(df.head(3))

# Printing First 3 rows of the standardized DataFrame
print("\nStandardized DataFrame:")
print(df_standardized.head(3))
```

```python
# Printing the scaled features as a dataframe
print(df_standardized)
```

```python
# show the correlation between columns
# Calculating the correlation matrix for Standardized Columns
correlation_matrix = df_standardized.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# perform PCA with no reducing(include all columns)
# PCA goes here.
from sklearn.decomposition import PCA
pca = PCA()
df_pca = pca.fit_transform(X_standardized)
```

```python
# Display the first 10 lines of the PCA output.
pca_output = pca.transform(X_standardized)
print("PCA Output (First 10 Lines):")
print(pca_output[:10])
```

```python
# Draw a bar chart showing the explained_variance of all PC's
# Calculating the explained variance ratios for all PCs
import matplotlib.pyplot as plt
variance_ratios = pca.explained_variance_ratio_
# Creating a bar chart for explained variance
plt.figure(figsize=(10, 6))
plt.bar(range(len(variance_ratios)), variance_ratios)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance of Principal Components')
plt.show()
```

```python
# Display the explained_variance_ for the model.
explained_variance = pca.explained_variance_
# Printing the explained variance for the model
print("Explained Variance for our Model:")
print(explained_variance)
```

```python
# Display scatter plots comparing the first, second, and third most important
# principal components in the PCA model
pc1 = df_pca[0]
pc2 = df_pca[1]
pc3 = df_pca[2]

# Create scatter plots for the first, second, and third principal components
plt.figure(figsize=(18, 6))

# Scatter plot for PC1 vs PC2
plt.subplot(131)
plt.scatter(pc1, pc2, c=Y[:len(pc1)], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PC1 vs PC2')

# Scatter plot for PC1 vs PC3
plt.subplot(132)
plt.scatter(pc1, pc3, c=Y[:len(pc1)], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 3')
plt.title('PC1 vs PC3')

# Scatter plot for PC2 vs PC3
plt.subplot(133)
plt.scatter(pc2, pc3, c=Y[:len(pc1)], cmap='viridis')
plt.xlabel('Principal Component 2')
plt.ylabel('Principal Component 3')
plt.title('PC2 vs PC3')

plt.tight_layout()
plt.show()
```

```python
# Perform a PCA using only the top 10 components (n_components=10)
# Code goes here.
# Performing PCA with n_components=10
pca_10 = PCA(n_components=10)
pca_output_10 = pca_10.fit_transform(X_standardized)
# Transforming the data using the top 10 components
pca_output_10 = pca.transform(X_standardized)
```

```python
# Displaying the first 10 lines of the PCA
print("PCA Output (First 10 Lines with Top 10 Components):")
print(pca_output_10[:10])
```

```python
# Draw a bar chart showing the explained_variance of the top 10 PC's
# Get the explained variance for the top 10 components
explained_variance_10 = pca_10.explained_variance_ratio_
# Creating a bar chart for explained variance of the top 10 components
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_10) + 1), explained_variance_10, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance of Top 10 Principal Components')
plt.show()
```

```python
# Printing the explained_variance_ for the model.
explained_variance_model = pca_10.explained_variance_
print("Explained Variance for the Model with Top 10 Components:")
print(explained_variance_model)
```

```python
# Display scatter plots comparing the first, second, and third most important
# principal components in the smaller PCA model
# Extract the first three components from the PCA results
pc1_10 = pca_output_10[:, 0]
pc2_10 = pca_output_10[:, 1]
pc3_10 = pca_output_10[:, 2]

# Creating scatter plots for the first, second, and third principal components in the smaller PCA model
plt.figure(figsize=(18, 6))

# Scatter plot for PC1 vs PC2
plt.subplot(131)
plt.scatter(pc1_10, pc2_10, c=Y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PC1 vs PC2 (Top 10 Components)')

# Scatter plot for PC1 vs PC3
plt.subplot(132)
plt.scatter(pc1_10, pc3_10, c=Y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 3')
plt.title('PC1 vs PC3 (Top 10 Components)')

# Scatter plot for PC2 vs PC3
plt.subplot(133)
plt.scatter(pc2_10, pc3_10, c=Y, cmap='viridis')
plt.xlabel('Principal Component 2')
plt.ylabel('Principal Component 3')
plt.title('PC2 vs PC3 (Top 10 Components)')

plt.tight_layout()
plt.show()
```

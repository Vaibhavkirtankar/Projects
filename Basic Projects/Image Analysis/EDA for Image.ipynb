```python
#Compute edge of the below given image using Prewitt and Sobel edge detectors. Compare
#the performance of both the detectors and show the output.
```

```python
# Compute the edges using perwitt and Sobel edge detectors
from skimage import filters
prewitt_edges = filters.prewitt(img3)
sobel_edges = filters.sobel(img3)

fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4),
                               sharex=True, sharey=True)
ax0.imshow(prewitt_edges, cmap=plt.cm.gray)
ax0.set_title('Prewitt edges')

ax1.imshow(sobel_edges, cmap=plt.cm.gray)
ax1.set_title('Sobel edges')

for ax in (ax0, ax1):
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Define range of scales to apply
scales = [3, 5, 7, 9]

# Apply LOG filter for each scale and plot results
fig, axs = plt.subplots(nrows=1, ncols=len(scales)+1, figsize=(16, 6))
axs[0].imshow(img3, cmap='gray')
axs[0].set_title('Original Image')
for i, scale in enumerate(scales):
    # Apply Gaussian blur
    img_blur = cv2.GaussianBlur(img3, (scale, scale), 0)
    # Apply Laplacian of Gaussian filter
    img_log = cv2.Laplacian(img_blur, cv2.CV_64F)
    # Normalize output to range [0, 255]
    img_log = cv2.normalize(img_log, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    axs[i+1].imshow(img_log, cmap='gray')
    axs[i+1].set_title('Scale = {}'.format(scale))

# Show plot
plt.show()
```

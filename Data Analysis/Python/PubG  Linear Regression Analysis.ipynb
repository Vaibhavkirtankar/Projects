```python
# 10. Take "Kills" as the response variable and "Ln Match Kill Rank" as the predictor.
import statsmodels.api as sm
X = df[['Ln Match Kill Rank']]
y = df['Kills']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 11.The plot for the above simple linear regression model.
import matplotlib.pyplot as plt
plt.scatter(X['Ln Match Kill Rank'], y, label="Data Points")
plt.xlabel("Ln Match Kill Rank")
plt.ylabel("Kills")
plt.legend()
plt.show()
```

```python
# 13. Using 'Winplaceperc' as the response variable, perform a full linear regression model
import statsmodels.api as sm
X = df.drop(columns=['Winplaceperc'])
y = df['Winplaceperc']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 14. Perform a 4-feature multiple linear regression using Winplaceperc as the response variable
# and "Kills", "Walk distance", "Matchkillrank", and "Boosts" as the predictors.
predictor_cols = ["Kills", "Walkdistance", "Matchkillrank", "Boosts"]
X = df[predictor_cols]
y = df['Winplaceperc']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 18. Display the partial regression plots on the 4-feature multiple linear regression

# Create partial regression plots
sns.set(style="whitegrid")
sns.mpl.rc("figure", figsize=(12, 6))

# Loop through each predictor variable
for i, predictor in enumerate(predictor_cols):
    plt.subplot(2, 2, i + 1)
    sns.residplot(x=model.fittedvalues, y=y - model.fittedvalues)
    plt.title(f"Partial Regression Plot\nfor {predictor}")
    plt.xlabel("Fitted Values")
    plt.ylabel(predictor)

plt.tight_layout()
plt.show()
```

```python
# 19. Draw the Q-Q plot for the 4-feature multiple linear regression

residuals = model.resid
sm.qqplot(residuals)
plt.title("Q-Q Plot")
plt.show()
```
